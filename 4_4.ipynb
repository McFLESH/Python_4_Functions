```python
# Объявите функцию с именем is_long, которая принимает отзыв в виде строки (в качестве аргумента)
# и возвращает False, если длина строки меньше десяти символов. Иначе возвращается значение True.
def is_long(stroka):
    kolvo = len(stroka)
    if kolvo >= 10:
        return True
    else:
        return False
```
