```python
# На вход программа должна получать пароль в виде строки формата "00#000#0",
# где числа вместо нуля будут подбираться вами самостоятельно. Строка имеет такой и только такой вид,
# то есть число a гарантированно является двузначным, число b гарантированно трехзначное, число c - однозначное.
# Символ # - это разделитель, который обязательно должен быть в строке, но не является символом для вычислений.

def funk_a(a):
    a = int(a)
    for i in range(2, a):
        if a % i == 0:
            return False
    return True
def funk_b(b):
    if b[0] == b[-1]:
        return True
    else:
        return False
def funk_c(c):
    c = int(c)
    if c % 2 == 0:
        return True
    else:
        return False
def check_password(password):
    spisok = password.split("#")
    if funk_a(spisok[0]) and funk_b(spisok[1]) and funk_c(spisok[2]) == True:
        return f'Пароль {password} подходит под условия'
    else:
        return f'Пароль {password} не подходит под условия'
print(check_password(input()))
```
