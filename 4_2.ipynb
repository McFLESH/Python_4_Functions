```python
# Напишите функцию summa_n, которая принимает одно целое положительное число t
# и находит сумму всех чисел от 1 до t включительно.

def summa_n(t):
    S = 0
    i = 1
    while i <= t:
        S = S + i
        i = i + 1
    return f"Я знаю, что сумма чисел от 1 до {t} равна {S}"
```
