```python
# Объявите функцию analyze_prices, которая принимает ряд цен на товары в виде строки,
# находит максимальное, минимальное значения и сумму значений этих цен
def analyze_prices(stroka):
    spisok = stroka.split()
    spisok = list(map(int, spisok))
    v_min = min(spisok)
    v_max = max(spisok)
    v_sum = sum(spisok)
    return f'Min = {v_min}, max = {v_max}, sum = {v_sum}'
```
